```python
import os
import requests

# Đảm bảo đang ở trong thư mục yolov7
if os.path.basename(os.getcwd()) != 'yolov7':
    %cd yolov7

url = "https://github.com/WongKinYiu/yolov7/releases/download/v0.1/yolov7-tiny.pt"
filename = "yolov7-tiny.pt"

# Tải file
response = requests.get(url)

# Kiểm tra nếu tải thành công (status code 200)
if response.status_code == 200:
    with open(filename, "wb") as f:
        f.write(response.content)
    print("yolov7-tiny.pt")
else:
    print(f"{response.status_code}")

# Kiểm tra dung lượng file để chắc chắn không bị lỗi 0KB
file_size = os.path.getsize(filename)
print(f"Kích thước file: {file_size / 1024 / 1024:.2f} MB")



```
